```python
openai = pd.read_csv('news_openai_final.csv')
roberta = pd.read_csv('prebuilt_rob_sentiment.csv')
finetuned_roberta = pd.read_csv('news_finetuned_roberta.csv')
```

```python
openai['Publishing Time'] = pd.to_datetime(openai['Publishing Time'])
roberta['Publishing Time'] = pd.to_datetime(roberta['Publishing Time'])
finetuned_roberta['Publishing Time'] = pd.to_datetime(finetuned_roberta['Publishing Time'])
```

```python
filtered_roberta = roberta[roberta['Publishing Time'].isin(openai['Publishing Time'])]
```

```python
filtered_sorted_roberta = filtered_roberta.sort_values(by='Publishing Time', ascending=True)
```

```python
sentiment_mapping = {
    'positive': 1.0,
    'negative': -1.0,
    'neutral': 0.0
}

filtered_sorted_roberta['rob_sentiment'] = filtered_sorted_roberta['rob_sentiment'].map(sentiment_mapping)
filtered_sorted_roberta['fin_sentiment'] = filtered_sorted_roberta['fin_sentiment'].map(sentiment_mapping)
```

```python
filtered_sorted_roberta.head(5)
```

```python
openai_columns = ['Publishing Time', 'openai_sentiment']
roberta_columns = ['Publishing Time', 'rob_sentiment', 'fin_sentiment']
finetuned_roberta_columns = ['Publishing Time', 'finetune_roberta_sentiment']
```

```python
merged_temp = pd.merge(openai[openai_columns], filtered_sorted_roberta[roberta_columns], on='Publishing Time')
merged_df = pd.merge(merged_temp, finetuned_roberta[finetuned_roberta_columns], on='Publishing Time')
```

```python
merged_df
```

```python
openai_roberta = merged_df[merged_df['openai_sentiment'] == merged_df['rob_sentiment']].shape[0]
openai_finvader = merged_df[merged_df['openai_sentiment'] == merged_df['fin_sentiment']].shape[0]
roberta_finvader = merged_df[merged_df['rob_sentiment'] == merged_df['fin_sentiment']].shape[0]
finetune_roberta_base = merged_df[merged_df['finetune_roberta_sentiment'] == merged_df['rob_sentiment']].shape[0]
finetune_roberta_openai = merged_df[merged_df['finetune_roberta_sentiment'] == merged_df['openai_sentiment']].shape[0]
finetune_roberta_finvader = merged_df[merged_df['finetune_roberta_sentiment'] == merged_df['fin_sentiment']].shape[0]
total_rows = merged_df.shape[0]
```

```python
percent_openai_roberta = (openai_roberta / total_rows) * 100
percent_openai_finvader = (openai_finvader / total_rows) * 100
percent_roberta_finvader = (roberta_finvader / total_rows) * 100

percent_finetune_roberta_base = (finetune_roberta_base / total_rows) * 100
percent_finetune_roberta_openai = (finetune_roberta_openai / total_rows) * 100
percent_finetune_roberta_finvader = (finetune_roberta_finvader / total_rows) * 100
```

```python
print(f"Percentage of matching sentiments openai vs roberta: {percent_openai_roberta:.2f}%")
print(f"Percentage of matching sentiments openai vs finvader: {percent_openai_finvader:.2f}%")
print(f"Percentage of matching sentiments roberta vs finvader: {percent_roberta_finvader:.2f}%")

print(f"Percentage of matching sentiments finetuned roberta vs roberta base: {percent_finetune_roberta_base:.2f}%")
print(f"Percentage of matching sentiments finetuned roberta vs openai: {percent_finetune_roberta_openai:.2f}%")
print(f"Percentage of matching sentiments finetuned roberta vs finvader: {percent_finetune_roberta_finvader:.2f}%")
```
